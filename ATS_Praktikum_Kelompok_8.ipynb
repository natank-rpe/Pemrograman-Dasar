```python
# Soal Nomor 3 - Looping for

# Data hasil pembacaan vibrasi dari 5 unit mesin
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

# Loop untuk memeriksa setiap nilai vibrasi
for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i+1} vibrasi tinggi!")
```

```python
# Soal Nomor 5 - Looping while

# Nilai awal suhu
suhu_saat_ini = 60

# Loop while untuk menaikkan suhu hingga mencapai 85°C
while suhu_saat_ini < 85:
    print(f"Suhu saat ini: {suhu_saat_ini}°C")
    suhu_saat_ini += 5

print("Target suhu 85°C tercapai!")
```

```python
# Soal Nomor 7 - Percabangan / Branching

# Input pengguna untuk memasukkan beban terpakai dan kapasitas maksimum
kapasitas_maks = float(input("Masukkan kapasitas maksimum (Watt): "))
beban_terpakai = float(input("Masukkan beban terpakai (Watt): "))

# Menghitung persentasi pemakaian
persen = (beban_terpakai / kapasitas_maks) * 100

# Percabangan untuk mengklafisikasikan beban listrik berdasarkan pemakaian
if persen <= 60:
  status = "Beban Ringan"
elif persen <= 85:
  status = "Beban Sedang"
elif persen > 85:
  status = "Beban Berat"

# Menampilkan hasil
print(f"Persentasi pemakaian: {persen:.2f}%")
print(f"Status: {status}")
```

```python
# Soal Nomor 9 - Looping + List

# Data suhu pada 5 titik di boiler
suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

# Loop untuk menampilkan setiap nilai suhu
for i in range(len(suhu_boiler)):
    print(f"Titik {i+1}: {suhu_boiler[i]}°C")
    if suhu_boiler[i] > 100:
        print(f"Peringatan: Suhu di titik {i+1} melebihi 100°C!")
    print()
```

```python
# CHALLENGE
# Monitoring Sederhana Parameter pada Boiler Pembangkit Listrik

def check_status(suhu, tekanan, vibrasi):
    if suhu >= 520 or tekanan >= 150:
        return "ALERT: SHUTDOWN"
    if suhu >= 500 or tekanan >= 140 or vibrasi > 5:
        return "Warning"
    return "Normal"

def main():
    while True:
        jumlah = int(input("Masukkan jumlah unit yang akan diperiksa: "))
        for i in range(1, jumlah + 1):
            print(f"\n-- Unit {i} --")
            suhu = float(input("Suhu boiler (°C): "))
            tekanan = float(input("Tekanan uap (bar): "))
            vibrasi = float(input("Vibrasi (mm/s): "))

            status = check_status(suhu, tekanan, vibrasi)

            print(f"\nHasil Unit {i}:")
            print(f"  Suhu   : {suhu:.1f} °C")
            print(f"  Tekanan: {tekanan:.1f} bar")
            print(f"  Vibrasi: {vibrasi:.2f} mm/s")
            print(f"  Status : {status}")
            print()  # spasi antar unit

        lagi = input("Ulangi pemantauan? (y/n): ").strip().lower()
        if lagi != 'y':
            print("Selesai.")
            break

if __name__ == "__main__":
    main()
```
