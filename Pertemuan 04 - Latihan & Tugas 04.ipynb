```python
# Latihan Nomor 1
# Program Menentukan Kelipatan 5

angka = int(input("Masukkan sebuah angka: "))

if angka % 5 == 0:
  print(angka, "Merupakan Kelipatan 5")
else:
  print(angka, "Bukan Kelipatan 5")
```

```python
# Latihan Nomor 2
# Program Menentukan Tahun Kabisat

tahun = int(input("Masukkan Tahun: "))

if tahun % 400 == 0:
  print(f"Tahun {tahun} adalah Tahun Kabisat")
elif tahun % 4 == 0 and tahun % 100 != 0:
  print(f"Tahun {tahun} adalah Tahun Kabisat")
else:
  print(f"Tahun {tahun} bukan Tahun Kabisat")
```

```python
# Latihan Nomor 3
# Program Soal Matematika dengan Kondisi If-Elif-Else

print ("Pilih Operasi Matematika:")
print ("1. Penjumlahan")
print ("2. Pengurangan")
print ("3. Perkalian")
print ("4. Pembagian")

pilihan = int(input("Masukkan pilihan operasi (1-4): "))
a = float(input("Masukkan angka pertama: "))
b = float(input("Masukkan angka kedua: "))

if pilihan == 1:
  hasil = a + b
  print("Hasil penjumlahan =", hasil)
elif pilihan == 2:
  hasil = a - b
  print("Hasil pengurangan =", hasil)
elif pilihan == 3:
  hasil = a * b
  print("Hasil perkalian =", hasil)
elif pilihan == 4:
  if b != 0:
    hasil = a / b
    print("Hasil pembagian =", hasil)
  else:
    print("Error: pembagian dengan nol tidak diperbolehkan!.")
else:
  print("Input salah. Silah pilih angka 1-4.")
```

```python
# Latihan Nomor 4
# Program Menentukan Bulan

bulan = int(input("Masukkan angka bulan (1-12): "))

daftar_bulan = ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli",
                "Agustus", "September", "Oktober", "November", "Desember"]
if 1 <= bulan <= 12:
  print(f"Bulan ke-{bulan} adalah {daftar_bulan[bulan-1]}")
else:
  print("Input salah")
```

```python
# Latihan Nomor 5
# Program Menampilkan Salam Sesuai Jam

jam = int(input("Masukkan jam (0–23): "))

if jam >= 0 and jam <= 11:
    print("Selamat Pagi")
elif jam >= 12 and jam <= 15:
    print("Selamat Siang")
elif jam >= 16 and jam <= 18:
    print("Selamat Sore")
elif jam >= 19 and jam <= 23:
    print("Selamat Malam")
else:
    print("Jam tidak valid")
```

```python
# Tugas Nomor 1
# Program Kasir Dengan 3 Kategori Diskon

print("=" * 30)
print("SELAMAT DATANG DI KAFE GEMINI")
print("=" * 30)
print("Menu Makanan & Minuman:")
print("1. Kopi Hitam   - Rp 12.000")
print("2. Teh Manis    - Rp 8.000")
print("3. Roti Bakar   - Rp 15.000")
print("4. Mie Instan   - Rp 10.000")
print("=" * 30)

pilihan = int(input("Silakan masukkan nomor pilihan Anda (1-4): "))

nama_pesanan = ""
harga = 0

# Menentukan harga berdasarkan pilihan
if pilihan == 1:
    nama_pesanan = "Kopi Hitam"
    harga = 12000
elif pilihan == 2:
    nama_pesanan = "Teh Manis"
    harga = 8000
elif pilihan == 3:
    nama_pesanan = "Roti Bakar"
    harga = 15000
elif pilihan == 4:
    nama_pesanan = "Mie Instan"
    harga = 10000
else:
    print("Maaf, pilihan menu tidak valid. Silakan jalankan ulang program.")
    exit()

print(f"\nAnda memesan: {nama_pesanan} dengan harga Rp {harga}")

# Memasukkan jumlah pesanan
jumlah = int(input("Masukkan jumlah pesanan: "))
total = harga * jumlah

# Menentukan kategori diskon
if total >= 100000:
    diskon = 0.70  # 70%
elif total >= 50000:
    diskon = 0.50  # 50%
elif total >= 20000:
    diskon = 0.30  # 30%
else:
    diskon = 0     # Tidak ada diskon

# Menghitung total setelah diskon
potongan = total * diskon
total_bayar = total - potongan

print("-" * 30)
print(f"Total Belanja  : Rp {total}")
print(f"Diskon         : {int(diskon * 100)}%")
print(f"Potongan Harga : Rp {int(potongan)}")
print(f"Total Bayar    : Rp {int(total_bayar)}")
print("-" * 30)

# Pembayaran
uang_dibayar = int(input("Masukkan jumlah uang Anda: Rp "))
if uang_dibayar >= total_bayar:
    kembali = uang_dibayar - total_bayar
    print(f"Kembalian Anda : Rp {int(kembali)}")
    print("Terima kasih telah berbelanja di Kafe Gemini!")
else:
    kurang = total_bayar - uang_dibayar
    print(f"Maaf, uang Anda kurang Rp {int(kurang)}")
```
