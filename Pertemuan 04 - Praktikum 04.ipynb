```python
# IF (Percabangan Tunggal)
# Program untuk mengecek bonus dan diskon

total_belanja = input("Total belanja: Rp ")

# Jumlah yang harus dibayar adalah berapa total belanjaanya
# Tapi kalau dapat dikon akan berkurang
bayar = int(total_belanja)

# Jika dia belanja di atas 100rb maka berikan bonus dan diskon
if int(total_belanja) > 100000:
  print("Selamat Karena Belanja > Rp. 100.000 anda mendapat Voucher Makan ")
  print("dan Diskon 5%")

# Hitung diskonnya
diskon = int(total_belanja) * 5/100 #5%
bayar = int(total_belanja) - diskon

# Cetak struk
print("Total yang harus dibayar: Rp %s" % bayar)
print("Terima kasih sudah berbelanja di Toko Kami")
print("Kami tunggu kunjungan anda selanjutnya.")
```

```python
# IF-Else

total_belanja = 21000

if total_belanja > 20000:
  diskon = total_belanja * 0.10
  total_bayar = total_belanja - diskon
  print(f"Anda mendapatkan diskon! Total bayar: Rp {total_bayar}")
else:
  print(f"Total bayar: Rp {total_belanja}")
```

```python
# IF-Else

angka = int(input("Masukkan Bilangan: "))

if angka % 2 == 0:
  print("Bilangan Genap")
else:
  print("Bilangan Ganjil")
```

```python
# IF-Elif-Else

nilai = 80
if nilai >= 85:
  print("Luar biasa! Nilai A")
elif nilai >= 70:
  print("Bagus! Nilai B")
elif nilai >= 60:
  print("Cukup. Nilai C")
else:
  print("Nilai kurang, perlu belajar lebih giat")
```

```python
# IF-Elif-Else

# Menentukan jenis diskon berdasarkan hari
hari: str = "Sabtu"
if hari == "Sabtu" or hari == "Minggu":
  diskon: float = 0.20 # Diskon 20%
  print(f"Selamat! Anda mendapatkan diskon akhir pekan sebesar {diskon*100}%.")
elif hari == "Jumat":
  diskon: float = 0.15 # Diskon 15%
  print(f"Diskon spesial Jumat! Anda mendapat potongan {diskon*100}%.")
else:
  diskon: float = 0.05 # Diskon 5%
  print(f"Anda mendapatkan diskon hari biasa sebesar {diskon*100}%.")
```

```python
# IF-Elif-Else

rupiah = int(input("Masukkan jumlah uang (Rp): "))
print("1. Dollar (USD)")
print("2. Euro (EUR)")
print("3. Yen (JPY)")
pilih = int(input("Pilih konversi (1/2/3): "))

if pilih == 1:
  print("Dalam Dollar:", rupiah / 15000)
elif pilih == 2:
  print("Dalam Euro:", rupiah / 16000)
elif pilih == 3:
  print("Dalam Yen:", rupiah / 120)
else:
  print("Pilihan tidak valid")
```
