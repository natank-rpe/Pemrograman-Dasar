```python
print("--- Soal 1 ---")
# Mencetak bilangan ganjil 1 hingga 9
for angka in range(2, 51, 2):
    print(angka)
```

```python
print("--- Soal 2 ---")
nim_berhenti = 4232411001
input_pengguna = 0 # Inisialisasi awal

while input_pengguna != nim_berhenti:
    try:
        input_pengguna = int(input("Masukkan NIM (ketik NIM untuk berhenti): "))
    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")

print(f"Perulangan telah dihentikan karena NIM {nim_berhenti} dimasukkan.")
```

```python
print("--- Soal 3 ---")
harga_barang = [15000, 25000, 10000, 5000, 7000, 6000, 12000, 30000, 35000, 2000]
total_belanja = 0

for harga in harga_barang:
    total_belanja += harga

print("Total biaya belanja adalah:", total_belanja)
```

```python
print("--- Soal 4 ---")
percobaan_maksimal = 3
percobaan_saat_ini = 1
passWord = "IU CANTIK"

while percobaan_saat_ini <= percobaan_maksimal:
    print(f"Percobaan ke-{percobaan_saat_ini}...")
    input_user = input("Masukkan password: ")

    if input_user == passWord:
        print("Akses diterima.")
        break
    else:
        print("Password salah.")
        percobaan_saat_ini += 1

if percobaan_saat_ini > percobaan_maksimal:
    print("Kesempatan habis. Akses ditolak.")
```

```python
print("--- Soal 5 ---")
nama_kampus = "Politeknik Negeri Batam"
hasil_output = ""

for karakter in nama_kampus:
    hasil_output += karakter + "-"

print(hasil_output) # Output: P-R-A-K-T-I-K-U-M-
```

```python
print("--- Soal 6 ---")
# Modifikasi countdown dimulai dari 255
for i in range(255, 0, -1):
    print(i)
```

```python
print("--- Soal 7 ---")
kelurahan_batam = ["Teluk Tering", "Lubuk Baja", "Belian", "Sungai Panas"]

for i, nama in enumerate(kelurahan_batam):
    print(f"Kelurahan ke-{i + 1}: {nama}")
```

```python
print("--- Soal 8 ---")
data_angka = [10, 20, 30, -1, 0, 50, 60]

for angka in data_angka:
    if angka == 0:
        print(f"Angka: {angka}")
        print("Angka 0 ditemukan!")
        break

    print(f"Angka: {angka}")
```

```python
print("--- Soal 9 ---")
total = 0
bilangan = 1
batas = 20

while total <= batas:
    total += bilangan
    bilangan += 1

print("Total terakhir:", total)
```
