```python
# Jawaban Soal Nomor 5

# Mengecek kelipatan 3 (menggunakan for dan while)

batas = int(input("Masukkan batas angka: "))
angka = 1

while angka <= batas:
    if angka % 3 == 0:
        print(f"{angka} merupakan kelipatan 3")
    else:
        print(f"{angka} bukan kelipatan 3")

    angka += 1  # update agar while terus maju
```
