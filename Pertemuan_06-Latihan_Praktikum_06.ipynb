```python
# SOAL 2: Menghitung Volume Kubus
def hitung_volume_kubus(sisi):
    """Menghitung volume kubus dari panjang sisinya."""
    return sisi ** 3

# Pemanggilan fungsi kubus
panjang_sisi = 5
print(f"Volume kubus dengan sisi {panjang_sisi} adalah: {hitung_volume_kubus(panjang_sisi)}")


# Modifikasi: Fungsi menghitung volume balok
def hitung_volume_balok(panjang, lebar, tinggi):
    """Menghitung volume balok berdasarkan panjang, lebar, dan tinggi."""
    return panjang * lebar * tinggi

# Pemanggilan fungsi balok
p = 5
l = 3
t = 6
print(f"Volume balok dengan ukuran {p}x{l}x{t} adalah: {hitung_volume_balok(p, l, t)}")
```

```python
# SOAL 3: Menghitung Rata-rata dari List Nilai (dengan penalti)
def hitung_rata_rata(data_nilai):
    """Menghitung rata-rata dari list angka menggunakan perulangan for dan penalti nilai tidak valid."""
    total = 0

    for nilai in data_nilai:
        # Pengecekan nilai valid
        if nilai < 0 or nilai > 100:
            print(f"Nilai {nilai} tidak valid! Dikenakan penalti -2.")
            nilai = -2  # Penalti untuk nilai tidak valid
        total += nilai

    # Pembagian untuk rata-rata
    if len(data_nilai) > 0:
        return total / len(data_nilai)
    return 0


# Pemanggilan contoh:
nilai_praktikan = [90, 110, 95, -10, 100]
print(f"\nNilai rata-rata kelas adalah: {hitung_rata_rata(nilai_praktikan):.2f}")
```

```python
# SOAL 4: Sistem Penentuan Predikat Nilai
def tentukan_predikat(skor):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor numerik (0–100)."""

    # Cek tipe data agar hanya menerima angka
    if not isinstance(skor, (int, float)):
        return "Input tidak valid! Skor harus berupa angka."

    # Cek rentang skor
    if skor > 100 or skor < 0:
        return "Input di luar rentang nilai (0–100)."

    # Percabangan nilai
    if skor >= 85:
        return "A"
    elif skor >= 70:
        return "B"
    elif skor >= 60:
        return "C"
    elif skor >= 50:
        return "D"
    else:
        return "E"


# Pemanggilan:
print(f"Skor 75 mendapat predikat: {tentukan_predikat(75)}")
print(f"Skor 48 mendapat predikat: {tentukan_predikat(48)}")

# Contoh input tidak valid
print(f"\nSkor 'abc' mendapat predikat: {tentukan_predikat('abc')}")
print(f"Skor 120 mendapat predikat: {tentukan_predikat(120)}")
```

```python
# SOAL 5: Menghitung Faktorial dengan Loop While
def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan bulat positif n."""

    # Tangani kasus khusus
    if n < 0:
        return "Tidak dapat menghitung faktorial dari bilangan negatif."
    elif n == 0:
        return 1

    hasil = 1

    # Perulangan berjalan selama n lebih besar dari 1
    while n > 1:
        hasil *= n
        n -= 1

    return hasil


# Pemanggilan:
bilangan_n = 4
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")  # Output: 24

# Contoh tambahan
print(f"Faktorial dari 0 adalah: {hitung_faktorial(0)}")
print(f"Faktorial dari -8 adalah: {hitung_faktorial(-8)}")
```

```python
# SOAL 6: Analisis Data Suhu Ekstrem (for dan if/elif)
def analisis_suhu(data_suhu, batas_panas=35, batas_dingin=15):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu."""
    panas_count = 0
    dingin_count = 0

    # Perulangan untuk mengecek setiap data suhu
    for suhu in data_suhu:
        if suhu >= batas_panas:
            panas_count += 1
        elif suhu <= batas_dingin:
            dingin_count += 1

    return panas_count, dingin_count


# Pemanggilan:
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28, 37]
jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)

print(f"Jumlah hari Panas (≥35°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (≤15°C): {jumlah_dingin}")
```
