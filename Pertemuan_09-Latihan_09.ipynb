```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
natan_df = pd.read_csv('/content/products.csv')
```

```python
# PRAKTIKUM 1 - Statistik Deskriptif (describe(), mean(), mode())

# 1. Ringkasan statistik
print("=== Ringkasan Statistik ===")
print(natan_df.describe(include='all'))

# 2. Rata-rata kolom numerik
rata_rata_price = natan_df['price'].mean()
rata_rata_quantity = natan_df['quantity'].mean()

print("\nRata-rata price :", rata_rata_price)
print("Rata-rata quantity :", rata_rata_quantity)

# 3. Modus kolom kategorikal (contoh: warna/colour)
modus_colour = natan_df['colour'].mode()[0]
modus_product_type = natan_df['product_type'].mode()[0]

print("\nModus colour :", modus_colour)
print("Modus product_type :", modus_product_type)
```

```python
# PRAKTIKUM 2 - Visualisasi Distribusi (hist())

# Histogram price
natan_df['price'].hist()
plt.title("Distribusi Harga Produk")
plt.xlabel("Harga")
plt.ylabel("Frekuensi")
plt.show()

# Histogram quantity
natan_df['quantity'].hist()
plt.title("Distribusi Jumlah Stok (Quantity)")
plt.xlabel("Quantity")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# PRAKTIKUM 3 - Korelasi & Kovarian (corr(), cov())

# Korelasi
corr_matrix = natan_df[['price', 'quantity']].corr()
print("Korelasi:\n", corr_matrix)

# Kovarian
cov_matrix = natan_df[['price', 'quantity']].cov()
print("Kovarian:\n", cov_matrix)
```

```python
# PRAKTIKUM 4  Groupby (groupby(), mean(), agg())

# Menambah kolom kategori
natan_df['Kategori_Stok'] = natan_df['quantity'].apply(lambda x: 'Banyak' if x >= 60 else 'Sedikit')

# Rata-rata price per kategori stok
group_mean = natan_df.groupby('Kategori_Stok')['price'].mean()
group_mean

# Agresi lengkap
group_agg = natan_df.groupby('Kategori_Stok').agg({
    'price': ['mean', 'max'],
    'quantity': ['mean', 'count']
})
group_agg
```
